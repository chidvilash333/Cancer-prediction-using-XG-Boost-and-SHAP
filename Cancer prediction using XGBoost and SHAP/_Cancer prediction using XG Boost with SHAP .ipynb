```python
# Load dataset
df = pd.read_csv('data (3).csv')

# Display the first few rows of the dataset to understand its structure
df.head()
```

```python
# Separating the features and target variable
X = df.iloc[:, :-1]  # Features
y = df.iloc[:, -1]   # Target

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Displaying the shape of the training and test sets to ensure proper splitting
print(f"Training set shape: {X_train.shape}")
print(f"Test set shape: {X_test.shape}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import xgboost as xgb

# Assuming `y_train` and `y` are already defined as the target arrays for training and the entire dataset
# Find common classes between training and entire dataset
common_classes = set(y_train).intersection(set(y))

# Filter out rows that contain uncommon classes
mask = np.isin(y, list(common_classes))
X_filtered = X[mask]
y_filtered = y[mask]

# Encode labels if needed, assuming they are not already encoded
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y_filtered)

# Split the filtered data
X_train, X_test, y_train_encoded, y_test_encoded = train_test_split(X_filtered, y_encoded, test_size=0.3, random_state=42)

# Initialize the XGBoost classifier
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')

# Train the model
xgb_model.fit(X_train, y_train_encoded)

# Predict on the test set
y_pred = xgb_model.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test_encoded, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```
